# **Load Dataset**

```python
import pandas as pd
file_path = "/content/drive/MyDrive/Dataset/share of self.xlsx"
df = pd.read_excel(file_path)
df 
```

# **DATA PREPROCESSING**
Data preprosessing pada dataset ini untu transform hurf menjadi lower case agar lebih akurat setelah untuk analisis. 


*   Contoh Outlet Name : Toko Majapahit / toko maJapahIt diasumsikaan adalah outlet yang sama



```python
df = df.applymap(lambda s: s.lower() if isinstance(s, str) else s)
df
```

# **Cek data yang kosong/ null data**


```python
#Cek missing data
null_data= df.isnull().sum()
null_data

#remove missing data
df =df.dropna()
df
```

# **BRAND AND FACING**
Brand **Anlene** mendominasi penempatan pada rak tapi **produgen** sebaliknya

```python
# Group the data by brand and flavor and sum the facing
facing_by_brand= df.groupby(["BRAND"])["FACING"].sum()

# Sort the facing_by_brandDataFrame in descending order
facing_by_brand= facing_by_brand.sort_values(ascending=False)
facing_by_brand
```

```python
#Visualization
facing_by_brand = facing_by_brand.sort_values(ascending=True)
ax = facing_by_brand.plot.barh(figsize=(10, 6), color="green")
ax.set_title("Facing by Brand")
ax.set_xlabel("Facing")
ax.set_ylabel("Brand")

# Add the sum for each brand to the bars
for i, v in enumerate(facing_by_brand):
    ax.text(v + 1, i, str(v), color='black', fontweight='bold')

plt.show()
```

```python
# Group the data by brand and flavor and sum the facing
facing_by_brand_flavor = df.groupby(["BRAND", "FLAVOR"])["FACING"].sum()

# Sort the facing_by_brand_flavor DataFrame in descending order
facing_by_brand_flavor = facing_by_brand_flavor.sort_values(ascending=False)
facing_by_brand_flavor
```

```python
# Visualization
facing_by_brand_flavor = facing_by_brand_flavor.reset_index()
g = sns.FacetGrid(facing_by_brand_flavor, col="BRAND", col_wrap=5)

# Map a histogram
g.map(sns.histplot, "FACING", bins=10)
g.fig.suptitle("Facing by Brand and Flavor")
g.set_xlabels("Total Facing")
g.set_ylabels("Count")
plt.show()
```

# **FACING BY TYPE STORE AND STORE NAME**
Local supermarket dengan nama Luwes purwodadi menajdi store yang paling facing barang (dapat dilihat pada diagram dibawah)

```python
facing_by_type_store = df.groupby(["TYPE OUTLET", "STORE NAME"])["FACING"].sum()
facing_by_type_store = facing_by_type_store.sort_values(ascending=False)
facing_by_type_store 
```

```python
# Create a horizontal bar chart
plt.figure(figsize=(10, 8))
ax = facing_by_type_store.plot(kind='barh')
plt.title('Total Facing by Store')
plt.xlabel('Total Facing')
plt.ylabel('Store Name')

# Add the total facing for each store at the end of the bar
for i, v in enumerate(facing_by_type_store):
    ax.text(v + 100, i - 0.1, str(v), color='blue', fontweight='bold')

plt.tight_layout()
plt.show()
```

# **TOTAL FACING BY TYPE OTLET**
Didapati hasil analisa **local supermarket **mendominasi Facing produk dibandingkan **supermarket**

```python
# Group the data by type outlet and sum the facing
facing_by_outlet = df.groupby("TYPE OUTLET")["FACING"].sum()

# Create a pie chart
plt.pie(facing_by_outlet, labels=facing_by_outlet.index, autopct='%1.1f%%')

# Set the title
plt.title("Total Facing by Type Outlet")

# Show the plot
plt.show()
```

# **FACING BY REGION**
Dapat dilihat bahwa **Central Region** adalah daerah yang paling banyak di di distribuskan produk dan brand **anlene** denga flavor coklat mendominasi

```python
facing_by_region = df.groupby(["REGION", "BRAND","FLAVOR"])["FACING"].sum()
facing_by_region = facing_by_region.sort_values(ascending=False)
facing_by_region
```

```python
#Visualization
pivot_df = df.pivot_table(index='REGION', columns='BRAND', values='FACING', aggfunc='sum')
sorted_df = pivot_df.loc[:, pivot_df.sum().sort_values(ascending=False).index]

# Create a stacked horizontal bar chart
plt.figure(figsize=(10, 6))
sorted_df.plot(kind='barh', stacked=True)
plt.title('TOTAL FACING PER REGION')
plt.xlabel('Total Facing')
plt.ylabel('Region')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.tight_layout()
plt.show()
```

# **PRINCIPLE vs COMPETITOR**
Dapat dilihat bahwa Facing product yang dimiliki competitor lebih bnayak dibandingkan principle

```python
# Group the data by MANDATORY and calculate the total facing
facing_by_mandatory = df.groupby('MANDATORY')['FACING'].sum()

# Create a bar chart
facing_by_mandatory.plot(kind='bar', color=['red', 'green'])
plt.title('Total Facing by Mandatory')
plt.xlabel('Mandatory')
plt.ylabel('Total Facing')
plt.xticks(rotation=0)
plt.show()
```

# **SEBARAN PRODUCT COMPETITOR VS PRINCIPLE**
Dapat dilihat pada grafik dibawah facing competitor lebih tinggi karena terdapat banyak jenis produk yang di facing.

Dapat dilihat juga  facing brand anleine,dancow,frisian flag lebih tinggi dibangkan boneeto, haillo,etc

```python
# Group the data by BRAND and MANDATORY and calculate the total facing
facing_by_brand_mandatory = df.groupby(['BRAND', 'MANDATORY'])['FACING'].sum().unstack()

# Sort the values by descending order of total facing
facing_by_brand_mandatory = facing_by_brand_mandatory.loc[:, facing_by_brand_mandatory.sum().sort_values(ascending=False).index]

# Create a stacked bar chart
facing_by_brand_mandatory.plot(kind='bar', stacked=True)
plt.title('Total Facing by Brand and Mandatory')
plt.xlabel('Brand')
plt.ylabel('Total Facing')
plt.legend(title='Mandatory', loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.tight_layout()
plt.show()
```
