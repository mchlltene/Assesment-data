# **PERBANDINGAN BARANG YANG AKAN EXPIRED**
Asumsikan hari adala tanggal (2022-09-15) maka:
*   5.4% produk sudah expired
*   5.6% produk akan expired dalam 3 bulan
*   89% produk belum expire



```python
# calculate the sum of expired products and products that have not yet expired
expired_sum = df.loc[df['Days Until Expiration'] <= 0, 'QUANTITY'].sum()
not_expired_sum = df.loc[df['Days Until Expiration'] > 0, 'QUANTITY'].sum()
three_month_sum = df.loc[(df['Days Until Expiration'] > 0) & (df['Days Until Expiration'] <= 90), 'QUANTITY'].sum()

# create a pie chart
labels = ['Expired', 'Not Expired', 'Expires in 3 Months']
sizes = [expired_sum, not_expired_sum, three_month_sum]
explode = (0.1, 0, 0)  # explode the "Expired" slice

fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Quantity of Expired and Not Yet Expired Products')
plt.show()
```

# **lINE CHART UNTUK MENUNJUKAN EXPIRED DATE**

```python
# convert the 'EXPIRED DATE' column to datetime objects
df['EXPIRED DATE'] = pd.to_datetime(df['EXPIRED DATE'])

# create a new column for the month and year of expiration
df['Expiration Month'] = df['EXPIRED DATE'].dt.to_period('M')

# group the data by month and year of expiration and sum the 'QUANTITY' column
monthly_totals = df.groupby('Expiration Month')['QUANTITY'].sum()

# create a line chart to visualize the trend of expired products over time
monthly_totals.plot(kind='line', x='Expiration Month', y='QUANTITY')
plt.title('Total Expired Products by Month and Year')
plt.xlabel('Expiration Month')
plt.ylabel('Total Quantity')
plt.show()
```
